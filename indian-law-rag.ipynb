```python
#convert dataframe to dataset
from datasets import Dataset
ds = Dataset.from_pandas(df_law)
ds
```

```python
# Convert rows into Documents with metadata
from langchain.schema import Document
docs = []
for rows in ds:
  metadata = {
      'act_title': rows['act_title'],

      'section': rows['section']
    }
  docs.append(Document(page_content=rows['law'], metadata=metadata))
```

```python
docs[0]
```

```python
# page content could be very large. so we need to chunk it.
# Create a text splitter to divide the text into manageable chunks
# Each chunk has a maximum size of 2048 characters with a 512-character overlap
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=2048, chunk_overlap=512)
split_docs = splitter.split_documents(docs)
```

```python
# Optimization techniques
# Hybrid search

#create BM25 retriever
from langchain.retrievers import BM25Retriever, EnsembleRetriever

bm25retriever = BM25Retriever.from_documents(split_docs)
bm25retriever.k = 10

#combine them into Hybrid retriever
hybrid_retriever = EnsembleRetriever(
    retrievers = [bm25retriever, retriever_chromadb],
    weights = [0.4, 0.6]
)
```

```python
query = "What is punishment for person who commits murder?"
out_docs1 = hybrid_retriever.invoke(query)
out_docs1
```

```python
query = "What is the punishment for theft?"
out_docs2 = hybrid_retriever.invoke(query)
out_docs2
```

```python
# Add CrossEncoder reranker model from hugging face
from sentence_transformers import CrossEncoder

# Load reranker model
reranker = CrossEncoder("BAAI/bge-reranker-base")
```

```python
#score hybrid search results
# Create query-document pairs
pairs = [[query, doc.page_content] for doc in out_docs1]

# Predict relevance scores
scores = reranker.predict(pairs)

# Attach scores to metadata
for doc, score in zip(out_docs1, scores):
    doc.metadata["relevance_score"] = float(score)

# Sort documents by relevance
reranked_docs = sorted(out_docs1, key=lambda x: x.metadata["relevance_score"], reverse=True)
```

```python
reranked_docs
```

```python
# Initialize an empty DataFrame with specified columns
rerank_df = pd.DataFrame(columns=['Law', 'section', 'relevance_score'])


# Loop through the retrieved documents and populate the DataFrame
for doc in out_docs1:
    rerank_df = rerank_df._append(
        {
            'Law': doc.page_content,  # Extract the page content (text) from the document
            'section': doc.metadata['section'],  # Extract the source metadata
            'relevance_score': doc.metadata['relevance_score']  # Extract the relevance score
        },
        ignore_index=True  # Ensure the new row is appended without overwriting existing rows
    )

# Display the first 3 rows of the DataFrame
rerank_df
```

```python
rerank_df = rerank_df.sort_values(by = 'relevance_score', ascending = False)
rerank_df
```

```python
#RAG pipeline
from langchain.schema.output_parser import StrOutputParser
from langchain_core.runnables import RunnableParallel,RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
from langchain_cohere import ChatCohere
# Initialize an LLM instance using Cohere's "command-r" model
# The temperature parameter controls randomness in the generated responses; 0 ensures deterministic outputs
llm = ChatCohere(model="command-a-03-2025", temperature=0)

# Define a prompt template for generating answers based on a given context and question
prompt_str = """Answer the question below using the context:

Context: {context}

Question: {question}

Answer: """

# Create a ChatPromptTemplate from the string template, enabling dynamic input for context and question
prompt = ChatPromptTemplate.from_template(prompt_str)

# Create a retrieval pipeline to fetch relevant context and pass through the user's question
retrieval = RunnableParallel(
    {
        # Use the cohere rerank as a retriever to fetch relevant context for the question
        "context": lambda query: compression_retriever.get_relevant_documents(query),

        # Pass through the user's input question without modification
        "question": RunnablePassthrough()
    }
)

# Define an output parser to format the generated response into a string
output_parser = StrOutputParser()


# Create a processing chain that retrieves context, formats the prompt, generates an LLM response, and parses the output
chain = retrieval | prompt | llm | output_parser
```

```python
# Invoke the chain of components (retrieval, prompt generation, LLM processing, and output parsing)
# The question "What is YOLO?" is passed through the chain to generate the response
response = chain.invoke("What is punishment for murder?")

# Print the response generated by the chain
print(response)
```
